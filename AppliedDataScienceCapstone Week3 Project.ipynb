2. Read Wiki URL

```python
wiki_url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
page = urlopen(wiki_url).read().decode('utf-8')
soup = BeautifulSoup(page,'html.parser')
table = soup.body.table
```

```python
df = pd.read_html(str(table),header=0)[0]
df .head()
```

```python
df1 = df[df.Borough != 'Not assigned']
df1 = df1.sort_values(by=['Postal Code','Neighborhood'])
df1
```

# Task 2

```python
import geocoder

geo = geocoder.arcgis('{},Toronto, Ontario'.format('M5A'))
geo.latlng
```

```python
df1 = df1.reindex(columns = df1.columns.tolist() + ['Latitude','Longitude'])
df1.columns = df1.columns.str.replace(' ', '')
```

```python
for index, row  in df1.iterrows():
    row.PostalCode
    geo = geocoder.arcgis('{},Toronto, Ontario'.format(row.PostalCode))                                               
    df1.at[index,'Latitude'] = geo.latlng[0]
    df1.at[index,'Longitude'] = geo.latlng[1]

df1.head()
```

# Task 3

```python
address = 'Toronto, Ontario'

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
toronto_data = df1

# create map of Manhattan using latitude and longitude values
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=11)

# add markers to map
for lat, lng, label in zip(toronto_data['Latitude'], toronto_data['Longitude'], toronto_data['Neighborhood']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
